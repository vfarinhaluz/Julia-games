## Two-player games

### Prisoner's dilemma:
The code below allows us to specify any two-player game via its payoff table. It then creates a "NormalForm" object which has been defined in the backgroud and can be used for calculations and plots.

The default game is the prisonner's dilemma.

#### Feasible payoff set:

```julia
#The following matrix can be updated:
normalTable=[
    (5.,5.) (0,6.);
    (6.,0.) (1.,1.)
]

normalform=NormalForm(normalTable) #Constructs NormalForm object

p=plotFeasible2(normalform); #Code that plots all relevant values
```

#### Plotting individually rational set

```julia
minimax=miniMaxProfile(normalform, true) #Calculates minimax profile of any normal form

plotMinimaxIR(normalform)
```

#### Plotting feasible and IR payoff set

```julia
plotall2(normalform) #Code that plots all relevant values
```

### Random two-player game:
The following code randomly generates a random two-player game by drawing each payoff from a uniform distribution on [0,1] and then plots its connected payoffs.

```julia
normalform=randomNormalForm((2,2));

minimax=miniMaxProfile(normalform, true) #Calculates minimax profile of any normal form

plotall2(normalform) #Code that plots all relevant values
```
